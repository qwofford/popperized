```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Distribution-fitting method outline from http://www.insightsbot.com/blog/WEjdW/fitting-probability-distributions-with-python-part-1

def fit_dist(samples):
    dist_tests = ['norm', 'expon', 'pareto', 'rayleigh', 'genextreme']
    dist_results = []
    params = {}
    
    # For plotting
    _, ax = plt.subplots()
    hist,_,_ = ax.hist(samples, density=True, label='Actual Data')
    
    for dist_name in dist_tests:

        # Fit sample data to distribution
        dist = getattr(stats, dist_name)
        param = dist.fit(samples)
        print("%s params: "%dist_name, end='')
        print(param)
        params[dist_name] = param

        # Run the Kolmogorov-Smirnov test on the distribution
        D, p = stats.kstest(samples, dist_name, args = param, 
                            alternative='greater',
                           mode='asymp')

        xs = np.linspace(np.min(samples),np.max(samples),1000)
        if dist_name == "pareto":
            ys=dist.pdf(xs,param[0],loc=param[1],scale=param[2])
        elif dist_name == "genextreme":
            ys=dist.pdf(xs,param[0],loc=param[1],scale=param[2])
        else:
            ys=dist.pdf(xs,loc=param[0],scale=param[1])
        if np.max(ys) < np.max(hist)*5:
            ax.plot(xs,ys,label=dist_name)
        dist_results.append((dist_name, p))
    
    ax.legend(loc='best', shadow=True)
    plt.show()

    print("Kolmogorov-Smirnov test results")
    for result in dist_results:
        print("%15s\t: %E"%(result[0], result[1]))
    dist_guess, p = (max(dist_results, key=lambda item:item[1]))
    print()
    print("Distribution guess is %s with a p value of %.15f"%(dist_guess, p))
    print()
    return (dist_guess, params[dist_guess])
```

# Getting Samples from data generator

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data_generator/results/result.csv')

actual_means = []
expected_means = []
requested_means = []
actual_stds = []
expected_stds = []
requested_stds = []

for frame in df.groupby(['uniq_id']):
    sleep_actual = frame[1].get(['actual_sleep_nsec']).values
    actual_means.append(np.mean(sleep_actual))
    actual_stds.append(np.std(sleep_actual))
    
    sleep_expected = frame[1].get(['expected_sleep_nsec']).values
    expected_means.append(np.mean(sleep_expected))
    expected_stds.append(np.std(sleep_expected))
    
    requested_means.append(frame[1].get(['mean']).values[0,0])
    requested_stds.append(frame[1].get(['std']).values[0,0])

fig, ax = plt.subplots(1,1)
plt.yscale("log")
ax.plot(actual_means, 'r.', label="actual mean")
ax.plot(expected_means, 'b+', label="expected mean")
ax.plot(requested_means, 'g', label="requested mean")
ax.legend(loc='best',frameon=True)
plt.show()

fig, ax = plt.subplots(1,1)
plt.yscale("log")
ax.plot(actual_stds, 'r.', label="actual std")
ax.plot(expected_stds, 'b+', label="expected std")
ax.plot(requested_stds, 'g', label="requested std")
ax.legend(loc='best',frameon=True)
plt.show()
```

```python
# Load the generated values

df = pd.read_csv('../data_generator/results/result.csv')
for dist_runs in df.groupby(['distribution']):
    for run in dist_runs[1].groupby(['uniq_id']):
        print("-----------------------")
        print(run[1].get(['distribution']).values[0,0])
        print("Param 1: %f\tParam 2: %f"%(
            run[1].get(['mean']).values[0,0],
            run[1].get(['std']).values[0,0]
        ))
        
        samples = run[1].get(['actual_sleep_nsec']).values

        # Run the distribution test
        dist, params = fit_dist(samples)
```

# Generating Samples from NumPy

```python
scale = 1 # The scale of the Rayleigh distribution
n = 1000000 # The number of points
noise_ratio = 0 #0.0004 # The percentage of values that are just noise

# Generate values from distribution
samples = np.random.rayleigh(scale, int(n*(1.0-noise_ratio)))

# Add noise
noise = np.random.normal(np.mean(samples), np.std(samples), int(n*(noise_ratio)))
samples = np.append(samples, noise)

plt.hist(samples, bins=10, density=True)

# Fit to distribution
dist, params = fit_dist(samples)

print(params)

xs = np.linspace(0,6,1000)
plt.plot(xs, stats.rayleigh.pdf(xs, loc=params[0], scale=params[1]), 'r')
plt.show()
# stats.rayleigh.pdf(args = params)

# Find maxima
print("Max value from sampling: %f"%np.max(samples))
print("Min value from sampling: %f"%np.min(samples))
```

```python
# Fitting a Pareto Distribution to the data

num_samples = 20000
cutoff = 0.05

new_samples = np.random.rayleigh(scale=params[1], size=num_samples)

xs = np.linspace(0,6,1000)
plt.hist(new_samples, density=True)
plt.plot(xs, stats.rayleigh.pdf(xs, loc=params[0], scale=params[1]), 'r')
plt.show()

top = new_samples[np.argsort(new_samples)]
top = top[int(num_samples*(1-cutoff)):] - np.quantile(new_samples,(1-cutoff))
plt.hist(top, density=True)

xs = np.linspace(0,3,1000)
gp_param = stats.genpareto.fit(top, 0)
print(gp_param)
plt.plot(xs, stats.genpareto.pdf(xs, gp_param[0], gp_param[1], gp_param[2]))

plt.show()
```
